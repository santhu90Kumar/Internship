```python
# Comman function for init driver
def init_driver(url):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--disable-blink-features=AutomationControlled')
    chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
    chrome_options.add_experimental_option('useAutomationExtension', False)
    
    pathlib.Path('download').mkdir(parents=True, exist_ok=True) 
    
    preferences = {
        "download.default_directory": os.getcwd() + os.path.sep + '/download',
    }

    chrome_options.add_experimental_option('prefs', preferences)

    driver = webdriver.Chrome(executable_path=CHROMEDRIVER_PATH,options=chrome_options)
    driver.maximize_window()
    driver.get(url)
    return driver
```

```python
def wait_for_page_load(xpath_url):
    try:
        element_present = EC.presence_of_element_located((By.XPATH, xpath_url))
        WebDriverWait(driver, 5).until(element_present)
    except TimeoutException:
        pass
    except NoSuchElementException:
        print("Element not found")
```

# 1.Scrape the details of most viewed videos on YouTube from Wikipedia: 

Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/

You need to find following details:
A) Rank B) Name C) Artist D) Upload date E) Views

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
```

```python
driver = init_driver(url)
```

```python
all_box = driver.find_elements_by_xpath('//div[@id="mw-content-text"]//table[@class="wikitable sortable jquery-tablesorter"]//tbody/tr')
```

```python
all_data = []
for one_box in all_box[:30]:
    rank = one_box.find_elements_by_xpath('.//td')[0].text
    name = one_box.find_elements_by_xpath('.//td')[1].text
    artist = one_box.find_elements_by_xpath('.//td')[2].text
    views = one_box.find_elements_by_xpath('.//td')[3].text
    upload_date = one_box.find_elements_by_xpath('.//td')[4].text
    
    all_data.append({
        'rank' : rank,
        'name' : name,
        'artist' : artist,
        'views' : views,
        'upload_date' : upload_date,
    })

df = pd.DataFrame(all_data)
df
```

# 5. Scrape the details of trending repositories on Github.com.

Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver.get('https://github.com/trending')
```

```python
all_box_url = driver.find_elements_by_xpath('//article/h1/a')
```

```python
all_urls = []
for one_url in all_box_url:
    all_urls.append(one_url.get_attribute('href'))
```

```python
all_data = []
for one_url in all_urls:
    
    driver.get(one_url)

    repository_title = driver.find_element_by_xpath('//h1/strong').text
    if len(driver.find_elements_by_xpath('//div[@class="BorderGrid-cell"]/p')) != 0:
        repository_description = driver.find_element_by_xpath('//div[@class="BorderGrid-cell"]/p').text
    else:
        repository_description = 'NA'
    
    if len(driver.find_elements_by_xpath('//div[@class="BorderGrid-cell"]//a[contains(text(),"Contributors")]/span[@class="Counter"]')) != 0:
        contributors_count = driver.find_element_by_xpath('//div[@class="BorderGrid-cell"]//a[contains(text(),"Contributors")]/span[@class="Counter"]').text
    else:
        contributors_count = 'NA'
        
    languages_used_elemens = driver.find_elements_by_xpath('//a[contains(@data-ga-click,"Repository, language")]')
    
    language_used = []
    for lang in languages_used_elemens:
        language_used.append(lang.text.split('\n')[0])

    all_data.append({
        'repository_title': repository_title,
        'repository_description': repository_description,
        'contributors_count': contributors_count,
        'language_used': ', '.join(language_used),
    })
    
    
df = pd.DataFrame(all_data)
df
```

# 10. Details of Datasets from UCI machine learning repositories.

Url = https://archive.ics.uci.edu/
You have to find the following details:
    
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
driver.find_element_by_xpath('//b').click()
```

```python
all_boxes = driver.find_elements_by_xpath('//table[@cellpadding="5" and @border="1"]/tbody/tr')
```

```python
all_list = []
for i, one_box in enumerate(all_boxes):
    
    # Skiping Column
    
    if i == 0:
        continue
        
    dataset_name = one_box.find_elements_by_xpath('.//td')[0].text
    data_type = one_box.find_elements_by_xpath('.//td')[1].text
    task = one_box.find_elements_by_xpath('.//td')[2].text
    attribute_type = one_box.find_elements_by_xpath('.//td')[3].text
    no_of_instances = one_box.find_elements_by_xpath('.//td')[4].text
    no_of_attribute = one_box.find_elements_by_xpath('.//td')[5].text
    year = one_box.find_elements_by_xpath('.//td')[6].text


    
    all_list.append({
        'dataset_name':dataset_name ,
        'data_type':data_type ,
        'task':task ,
        'attribute_type':attribute_type ,
        'no_of_instances':no_of_instances ,
        'no_of_attribute':no_of_attribute ,
        'year':year ,
    })
    
    
df = pd.DataFrame(all_list)
df
```
