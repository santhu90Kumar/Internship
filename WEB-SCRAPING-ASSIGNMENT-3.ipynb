# Question No:01

1. Write a python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.
For e.g. If user input is ‚Äòguitar‚Äô. Then search for guitars.


2. In the above question, now scrape the following details of each product listed in first 3 pages 
of your search results and save it in a data frame and csv. In case if any product vertical has 
less than 3 pages in search results then scrape all the products available under that product 
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of 
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" 
and ‚ÄúProduct URL‚Äù. In case, if any of the details are missing for any of the product then 
replace it by ‚Äú-‚Äú.


```python
#importing libraries
import selenium
from selenium import webdriver
import pandas as pd
import time
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings("ignore")
```

```python
# Connect to web driver 
driver = webdriver.Chrome("chromedriver.exe")
```

```python
# 1.Getting url/web page.

url= "https://www.amazon.in/"
driver.get(url)
```

```python
# Creating a variable to save input variable.
product_name= input('Enter the product to be searched:')
```

```python
# Finding the web element for search bar.
# 2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúSkill,Designations,Companies‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field
search_job= driver.find_element_by_id('twotabsearchtextbox')

# write on search bar 'Data Analyst'
search_job.send_keys(product_name)

time.sleep(4)  

# 3.clicking the search button using XPATH function.
search_button= driver.find_element_by_id("nav-search-submit-button")
search_button.click()
```

```python
#empty Lists
#Lists
brand=[]
prod_name=[]
rating=[]
rating_num=[]
price= []
ret_ex=[]
availability=[]
prod_url=[]
```

```python

# Fetching urls to open each page.
urls =[]
for i in range(0,3): #upto 3 pages
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(j.get_attribute('href'))
    driver.find_element_by_xpath("//li[@class='a-last']/a").click()
    time.sleep(4)
```

```python
for i in urls[:150]: # each page contains 50 items, hence kept the number as 150 for 3 pages.
    driver.get(i)
    time.sleep(3)
    
    #fetching brand name
    try:
        brand_name=driver.find_element_by_xpath("//td[@class='a-span9']/span")
        brand.append(brand_name.text)
    except NoSuchElementException:
        brand.append('-')    
        
        
      #fetching product name
    try:
        product=driver.find_element_by_id("productTitle")
        prod_name.append(product.text)
    except NoSuchElementException:
        prod_name.append('-') 
        
        
      #fetching rating
    try:
        ratestar=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")#locating the rating
        rating.append(ratestar.text)#appending the ratings in Ratings list
    except NoSuchElementException:
        rating.append("No Rating") #appending the No rating if no rating is there
        
        
        
      #fetching rating number
    try:
        ratenum=driver.find_element_by_id("acrCustomerReviewText")
        rating_num.append(ratenum.text)
    except NoSuchElementException:
        rating_num.append('-') 
        
        
        
      #fetching Price
    try:
        prod_price=driver.find_element_by_id("priceblock_ourprice")
        price.append(prod_price.text)
    except NoSuchElementException:
        price.append('-') 
        
        
        
     #fetching Return/Exchange
    try:
        ret_exch=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        ret_ex.append(ret_exch.get_attribute("href"))
    except NoSuchElementException:
        ret_ex.append('-') 
        
           
        
       
     #fetching Availability
    try:
        availa=driver.find_element_by_id("availability")
        availability.append(availa.text)
    except NoSuchElementException:
        availability.append('-') 
        
        
        
           
     #fetching Product URL
    try:
        pro_url=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']/a")
        prod_url.append(pro_url.get_attribute("href"))
    except NoSuchElementException:
        prod_url.append('-')
```

```python
# Finally create a dataframe of the scraped data.
amazon_data =pd.DataFrame({})
amazon_data['Brand Name'] = brand
amazon_data['Product Name'] = prod_name
amazon_data['Rating'] = rating
amazon_data['Rating Number'] = rating_num
amazon_data['Price'] = price
amazon_data['Return/Exchange'] = ret_ex
amazon_data['Availability'] = availability
amazon_data['Product URL'] = prod_url

amazon_data
```

# Question No: 5. 
    
Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on 
google maps.


```python
#importing required libraries
import selenium
from selenium import webdriver
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import warnings
warnings.filterwarnings("ignore")
```

```python
# Connect to the web driver 
driver=webdriver.Chrome("chromedriver.exe")
```

```python
wait = WebDriverWait(driver, 10) # Waiting for chrome to launch
driver.get("https://www.google.com/maps") #google maps url
city= input("Enter the city: ") #Enter the city to search
wait.until(EC.element_to_be_clickable((By.ID, "searchboxinput"))).send_keys(city) #sending input city to search box
wait.until(EC.element_to_be_clickable((By.ID, "searchbox-searchbutton"))).click() #clicking search button.
time.sleep(5)

# ActionChains are a way to automate low-level interactions such as mouse movements, mouse button actions, keypress, and context menu interactions.
# right click operation and then perform
ActionChains(driver).move_to_element(driver.find_element_by_xpath("//html/body")).context_click().perform()


#copying first dropdown data from the right click menu dropdown.
print(wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "ul[role='menu']>li div div[class*='text']:nth-of-type(1)"))).text)
```
